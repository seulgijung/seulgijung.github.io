```python
import altair as alt
import pandas as pd

# data load and transpose (wide to long)
# index_col sets the first column(0) as the index
df = pd.read_csv("data/chart_1.csv", header=None, index_col=0).T
df.columns = ["year", "ratio"]  # reassuring line just in case
df["year"] = pd.to_datetime(df["year"], format="%Y")
df["ratio"] = df["ratio"].astype(float)
df["ratio_above"] = df["ratio"].clip(lower=105.22)  # create a upper base to fill area; lower base = 105.22

# area (fill only above 105.22)
area = alt.Chart(df).mark_area(
    color=alt.Gradient(
        gradient="linear",
        stops=[
            alt.GradientStop(color="white", offset=0),  # offset dictates the order of colors: 0 = start and 1 = end
            alt.GradientStop(color="#C23B22", offset=0.5),  # color when reaching 50%
            alt.GradientStop(color="#8e1d09", offset=1),
        ],
        x1=1, x2=1, y1=1, y2=0,     # a pair of y1 and y2 dictates the direction of the palette; 0 and 1 here are relative coordinates of the canvas, 0 = uppermost while 1 = bottommost; x1=1, x2=1: no gradient crosswise
    ),
).encode(
    alt.X("year:T", axis=alt.Axis(format="%Y")),
    alt.Y("ratio_above:Q", scale=alt.Scale(domain=[103, 117])),     # upper base
    alt.Y2(datum=105.22),   # lower base
)

# line 
line = alt.Chart(df).mark_line(color="#C23B22", strokeWidth=2).encode(
    alt.X("year:T"),
    alt.Y("ratio:Q"),
)

# dash line (y=105.22)
rule = alt.Chart(pd.DataFrame([{"y": 105.22}])).mark_rule(
    color="#968987", strokeWidth=1, strokeDash=[4, 4],  # strokeDash=[lnegth, gap]
).encode(
    alt.Y("y:Q"),   # y is numerical so quantative
)

chart_1 = (area + line + rule).properties(width=1600, height=550).configure_axis(grid=False)
chart_1.save("chart_1.svg")
```
